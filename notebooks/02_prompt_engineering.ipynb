```python
# %% Cell 1: Setup
import sys
sys.path.insert(0, "..")
from src.prompt_engineer import PromptEngineer

engineer = PromptEngineer(default_style="digital_art", default_quality="high")
```

```python
# %% Cell 2: Basic Prompt Building
result = engineer.build_prompt("a dragon flying over a castle")
print("POSITIVE:", result.positive)
print("\nNEGATIVE:", result.negative)
print(f"\nTokens: ~{result.tokens_estimated}")
```

```python
# %% Cell 3: Style Comparison
prompt = "portrait of an elf warrior"
for style in ["photorealistic", "anime", "oil_painting", "fantasy"]:
    r = engineer.build_prompt(prompt, style=style)
    print(f"\n{'='*50}")
    print(f"Style: {style}")
    print(f"Positive: {r.positive[:80]}...")
```

```python
# %% Cell 4: Quality Level Comparison
for quality in ["draft", "standard", "high", "ultra"]:
    r = engineer.build_prompt("a sunset", quality=quality)
    print(f"\n{quality.upper()}: {r.positive[:60]}...")
```

```python
# %% Cell 5: Emphasis Weighting
result = engineer.build_prompt(
    "a cat sitting on a windowsill, rain outside",
    emphasis={"rain outside": 1.5, "cat": 1.3}
)
print("Weighted prompt:", result.positive)
```

```python
# %% Cell 6: Prompt Analysis
prompts = [
    "cat",
    "a beautiful sunset over the ocean",
    "photorealistic portrait of a warrior, dramatic lighting, detailed armor, sharp focus, 8k",
]
for p in prompts:
    analysis = engineer.analyze_prompt(p)
    print(f"\n'{p[:40]}...' -> Score: {analysis['score']}/100 ({analysis['rating']})")
    for s in analysis["suggestions"]:
        print(f"  💡 {s}")
```

```python
# %% Cell 7: Enhancement
basic = "a forest"
enhanced = engineer.enhance_prompt(basic, ["lighting", "detail", "atmosphere", "depth"])
print(f"Basic: {basic}")
print(f"Enhanced: {enhanced}")
```

```python
# %% Cell 8: Variations
variations = engineer.create_variation_prompts(
    "a medieval town square",
    num_variations=4,
    variation_type="style"
)
for v in variations:
    print(f"\nStyle: {v.style}")
    print(f"Prompt: {v.positive[:60]}...")
```
